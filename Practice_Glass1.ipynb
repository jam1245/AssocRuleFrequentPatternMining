**Activity 1: ** Read the glass data file from **/datasets/glass/glass.txt**. Get the names of glass dataset columns from readme file in **/datasets/glass/header.txt** and assign the names to columns of glass_data

```r
glass_data <- read.table("../../../datasets/glass/glass.txt", header=F, sep=",")
head(glass_data)
```

### Data Processing

The data should be cleaned as it has unnecessary columns and numeric data should be converted to ordered factors.

**Activity 2: **  Remove the two attribute **Id** as it is just an identifier. All the other attributes are either numeric or integer. Map the attributes to ordinal attributes by building suitable categories like low, medium, high or very high.

```r
## Your code for activity 2 goes here
glass_data$Id <- NULL 
glass_data$Type_glass <- NULL 
glass_data$Ba <- NULL 
glass_data$Fe <- NULL 

summary(glass_data)
```

```r
glass_data$RI <- ordered(cut(glass_data$RI, c(1.51,1.515,1.518,1.52,1.53)), labels = c("low", "medium", "high","very high"))
glass_data$Na <- ordered(cut(glass_data$Na, c(10.73,12.91,13.3,13.82,17.38)), labels = c("low", "medium", "high","very high"))
glass_data$Mg <- ordered(cut(glass_data$Mg, c(0.0,2.115,3.48,3.6,4.48)), labels = c("low", "medium", "high","very high"))
glass_data$Al <- ordered(cut(glass_data$Al, c(0.290,1.19,1.445,1.6,3.4)), labels = c("low", "medium", "high","very high"))
glass_data$Si <- ordered(cut(glass_data$Si, c(69.81,72.28,72.79,73,75.39)), labels = c("low", "medium", "high","very high"))
glass_data$K <- ordered(cut(glass_data$K, c(0,0.122,0.555,0.61,6.2)), labels = c("low", "medium", "high","very high"))
glass_data$Ca <- ordered(cut(glass_data$Ca, c(5.43,8.2,8.6,9.17,16)), labels = c("low", "medium", "high","very high"))
```

```r
glass_data$Type_glass <- ordered(cut(glass_data$Type_glass, c(1.0,1.0,2.0,3.0,7.0)), labels = c("low", "medium", "high","very high"))
```

```r
glass_data$Fe <- ordered(cut(glass_data$Fe, c(0.0,0.0,0.570,0.1,0.5)), labels = c("low", "medium", "high","very high"))
```

```r
glass_data$Ba <- ordered(cut(glass_data$Ba, c(0.0,0.0,0.17,0.0,3.15)), labels = c("low", "medium", "high","very high"))

str(glass_data)
```

**Activity 3: ** Now, recode the data as a binary incidence matrix by coercing the data set to transactions. save the transactions to glass_trans variable.

```r
## Your code for activity 3 goes here
library("arules")
glass_trans <- as(glass_data, "transactions")
glass_trans
```

```r
rules <- apriori(glass_data, parameter=list(support=0.001, confidence=0.5))
```

**Activity 4: ** Draw a frequency plot for the generated transactions with a support of 0.1

```r
## Your code for activity 4 goes here
library("arulesViz")
itemFrequencyPlot(glass_trans, support = 0.1, cex.names=0.8)
```

** Activity 5:** Display the first three rules using inspect function.

**Activity 6: ** Draw a scatter plot for the generated transactions using measures 'support', 'lift' and use 'confidence' for shading.

```r
## Your code for activity 6 goes here
plot(rules, measure=c("support", "lift"), shading="confidence")
```
